```python
import torchvision
from torch import optim, nn
import matplotlib.pyplot as plt
```

```python
EPOCHS = 10
ITERS_PER_EPOCH = 50
model = torchvision.models.vgg16()
optimizer = optim.SGD(params=model.parameters(), lr=0.5)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer=optimizer, T_max=ITERS_PER_EPOCH)
```

```python
lr_ls = []
for epoch in range(EPOCHS):
    for iter in range(ITERS_PER_EPOCH):
        # train model
        optimizer.step()
        lr = scheduler.get_lr()
        lr_ls.append(lr)
        scheduler.step()
plt.plot(range(len(lr_ls)), lr_ls)
```
