```python
%matplotlib inline
import numpy as np
import torch
from mpl_toolkits import mplot3d
from d2l import torch as d2l
```

```python
def f(x):
    """ 风险函数 """
    return x * torch.cos(x * np.pi)

def g(x):
    """ 经验风险函数 """
    return f(x) + .2 * torch.cos(5 * x * np.pi)
```

```python
x = torch.arange(.5, 1.5, .01)
d2l.set_figsize((4.5, 2.5))
d2l.plot(x, [f(x), g(x)], 'x', 'risk')
```

```python
x = torch.arange(-1., 2., .01)
d2l.plot(x, [f(x), ], xlabel='x', ylabel='f(x)')
```

```python
x = torch.arange(-2., 2., .01)
d2l.plot(x, [x ** 3, ], xlabel='x', ylabel='f(x)')
```

```python
f = lambda x: .5 * x ** 2
g = lambda x: torch.cos(np.pi * x)
h = lambda x: torch.exp(.5 * x)

x, segment = torch.arange(-2, 2, .01), torch.tensor([-1.5, 1])
d2l.use_svg_display()
_, axes = d2l.plt.subplots(1, 3, figsize=(9, 3))
for ax, func in zip(axes, [f, g, h]):
    d2l.plot([x, segment], [func(x), func(segment)], axes=ax)
```

```python
f = lambda x: (x - 1) ** 2
d2l.set_figsize()
d2l.plot([x, segment], [f(x), f(segment)], 'x', 'f(x)')
```
